# Behavior Cloning

1. Generate expert data.
2. Create a superverised learning model from the data
3. Test the data

```python
# Generate expert data

import hw1files
import hw1files.tf_util
import hw1files.load_policy
from hw1files.run_expert import expert_actions

#expert_policy_file = 'hw1files/experts/Humanoid-v1.pkl' 
#envname = 'Humanoid-v2'
expert_policy_file = 'hw1files/experts/Hopper-v1.pkl' 
envname = 'Hopper-v2'
max_timesteps = None
num_rollouts = 20
render = False

data = expert_actions(expert_policy_file, envname, max_timesteps, num_rollouts, render)
```

```python

expert_observations = data['observations']
expert_actions = data['actions']

act_shape = expert_actions.shape
expert_actions = expert_actions.reshape([expert_actions.shape[0], expert_actions.shape[2]])

print(expert_observations.shape)
print(expert_actions.shape)

obs_len = len(expert_observations[0])
print(obs_len)

action_len = len(expert_actions[0])
print(action_len)
```

```python
# Create a neural network that takes in the right number of inputs and produces the right number of outputs
from keras.models import Sequential
from keras.layers import Dense

model = Sequential()
model.add(Dense(121, input_dim=obs_len, activation='relu'))
model.add(Dense(44, activation='relu'))
model.add(Dense(22, activation='relu'))
model.add(Dense(11, activation='relu'))
model.add(Dense(action_len, activation='sigmoid'))

model.compile(loss='mean_squared_error', optimizer='adam', metrics=['accuracy'])
model.fit(expert_observations, expert_actions, epochs=20, batch_size=10)
#model.evaluate(observations, actions, epo)
```

```python
# Run the supervised model and produce the reward function
import gym.spaces
import numpy as np
env = gym.make(envname)
max_steps = max_timesteps or env.spec.timestep_limit

returns = []
observations = []
actions = []
for i in range(num_rollouts):
    print('iter', i)
    obs = env.reset()
    done = False
    totalr = 0.
    steps = 0
    
    while not done:
        # action = policy_fn(obs[None,:])
        action = model.predict(obs.reshape(1, obs_len))
        observations.append(obs)
        actions.append(action)
        obs, r, done, _ = env.step(action)
        totalr += r
        steps += 1
        if render:
            env.render()
        if steps % 100 == 0: print("%i/%i"%(steps, max_steps))
        if steps >= max_steps:
            break
        returns.append(totalr)

print('returns', returns)
print('mean return', np.mean(returns))
print('std of return', np.std(returns))
```

The expert gets a reward of ~3778 but our behavior cloning system can only manage ~30-40. The standard deviation is also much higher than the expert suggesting that the rewards are more variable and possibly more to random chance than to a learned behavior. 

The simulation ends much earlier for the behavior cloning and that is possibly one of the reasons the scores are dramatically much lower.
